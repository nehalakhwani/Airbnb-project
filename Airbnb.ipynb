```python
import pandas as pd
df=pd.read_excel('AB_NYC_2019.xlsx( global airbnb).xlsx')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().mean()*100
```

```python
df['name'].fillna(df['name'].mode()[0],inplace = True)
```

```python
df['host_name'].fillna(df['host_name'].mode()[0],inplace = True)
```

```python
df['last_review'].fillna(df['last_review'].median(),inplace = True)
```

```python
df['reviews_per_month'].fillna(df['reviews_per_month'].mean(),inplace = True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.to_csv('cleaned_data.csv', index=False)
```
